**NE_Rainfall 2005-2016** 

**The** **Dataset**

The dataset used gives the month wise all India rainfall data. The sub-division wise rainfall and its departure from normal for each month and season has also been provided in the data.It contains the following columns:



*   Year: Year of data
*   Actual Rainfall:JUN  : Rainfall in NE India for the month of June
*   Actual Rainfall:JUL  : Rainfall in NE India for the month of July
*   Actual Rainfall:AUG  : Rainfall in NE India for the month of August
*   Actual Rainfall:SEPT : Rainfall in NE India for the month of September
*   Actual Rainfall:JUN-SEPT : Cumulative Rainfall in NE India for the period June to September
*   Departure Percentage: JUN : Departure Percentage of Rainfall in NE India for the month of June
*   Departure Percentage: JUL : Departure Percentage of Rainfall in NE India for the month of June
*   Departure Percentage: AUG : Departure Percentage of Rainfall in NE India for the month of June
*   Departure Percentage: SEPT : Departure Percentage of Rainfall in NE Indiafor the month of June
*   Departure Percentage: JUN-SEPT : Cumulative Departure Percentage of Rainfall in NE India for the period June to September
